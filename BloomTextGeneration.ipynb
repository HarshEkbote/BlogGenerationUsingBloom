```python
from transformers import AutoModelForCausalLM, AutoTokenizer, set_seed
```

```python
import  torch
torch.set_default_tensor_type(torch.cuda.FloatTensor)
```

```python
! nvidia-smi
```

```python
model=AutoModelForCausalLM.from_pretrained("bigscience/bloom-3b",use_cache=True)
tokenizer=AutoTokenizer.from_pretrained('bigscience/bloom-3b')
```

```python
set_seed(424242)
```

```python
model.__class__.__name__
```

```python
blogpost_title='Indian Classical Music'
```

```python
prompt=f'This is a post on {blogpost_title} \n {blogpost_title} '
```

```python
input_ids=tokenizer(prompt,return_tensors='pt').to(0)
```

```python
sample=model.generate(**input_ids,max_length=200,top_k=1,temperature=0.9,repetition_penalty=2.0)
```

```python
print(tokenizer.decode(sample[0],truncate_before_pattern=[r"\n\n^#","^'''","\n\n\n"]))
```

```python
sample=model.generate(**input_ids,max_length=200,num_beams=2,num_beam_groups=2,top_k=1,temperature=0.9,repetition_penalty=2.0,diversity_penalty=1.0)
```

```python
print(tokenizer.decode(sample[0],truncate_before_pattern=[r"\n\n^#","^'''","\n\n\n"]))
```
